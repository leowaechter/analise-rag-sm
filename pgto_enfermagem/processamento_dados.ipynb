```python
import pytesseract
from pdf2image import convert_from_path
import re 
import pandas as pd

pdf_path = "../pdfs/relatorio-anual-de-gestao-2024-rag.pdf"
numero_pagina = 13
imagens = convert_from_path(pdf_path, first_page=numero_pagina + 1, last_page=numero_pagina + 1)
imagem_pagina = imagens[0]

config_tesseract = '--psm 6'

texto_extraido = pytesseract.image_to_string(
    imagem_pagina, 
    lang='por', 
    config=config_tesseract
)

texto_corrigido = texto_extraido.replace('Parcela\nNovembro', 'Parcela Novembro')
linhas = texto_corrigido.strip().split('\n')
nomes_colunas = linhas[0].split()
import pandas as pd
import re

texto_corrigido = texto_extraido.replace('Parcela\nNovembro', 'Parcela Novembro')

nomes_colunas_corretos = [
    'MESES DE REFERÊNCIA',
    'SMS',
    'UPA',
    'SAMU',
    'GESTÃO DUPLA',
    'TOTAL'
]
# -----------------------------

linhas = texto_corrigido.strip().split('\n')
dados_processados = []
padrao_valor = r'R\$\d{1,3}(?:\.\d{3})*,\d{2}'

for linha in linhas[1:]:
    if 'TOTAL:' in linha or not linha.strip():
        continue

    valores = re.findall(padrao_valor, linha)
    
    if valores:
        primeiro_valor_pos = linha.find(valores[0])
        descricao = linha[:primeiro_valor_pos].strip()
        
        nova_linha = [descricao] + valores
        dados_processados.append(nova_linha)

df = pd.DataFrame(dados_processados, columns=nomes_colunas_corretos)
df.drop(df.index[13:15], inplace=True)
df
df.to_csv("../pgto_enfermagem/dados_pgto_enf.csv")
```
